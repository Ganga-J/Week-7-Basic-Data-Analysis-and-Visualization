Task 1: Load and Explore the Dataset
First, we'll import the necessary libraries, pandas and matplotlib. Then, we'll load the Iris dataset, which is a great choice for this exercise because it's clean and easy to work with.

```python
# Task 1: Load and Explore the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset using scikit-learn
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Display the first few rows
print("First 5 rows of the dataset:")
print(iris_df.head())

# Check data types and missing values
print("\nDataset Info:")
iris_df.info()
```

Task 2: Basic Data Analysis
Next, we'll compute basic descriptive statistics for the numerical columns and then perform a group-by analysis on the categorical 'species' column.

```python
# Compute basic statistics for numerical columns
print("Descriptive Statistics:")
print(iris_df.describe())

# Group by 'species' and compute the mean of all numerical columns for each species
print("\nMean of numerical columns grouped by species:")
species_mean = iris_df.groupby('species').mean()
print(species_mean)
```

Task 3: Data Visualization
Now, we'll create four different plots to visualize the data, customizing them with titles and labels to make them more informative

```python
# Set the style for the plots
sns.set_style("whitegrid")

# 1. Bar Chart: Average petal length per species
plt.figure(figsize=(8, 6))
species_mean['petal length (cm)'].plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Average Petal Length per Iris Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.xticks(rotation=0)
plt.show()

# 2. Histogram: Distribution of sepal length
plt.figure(figsize=(8, 6))
plt.hist(iris_df['sepal length (cm)'], bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# 3. Scatter Plot: Relationship between sepal length and petal length
plt.figure(figsize=(10, 8))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df, s=80)
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()

# 4. Line Chart: This dataset does not have a time-series component,
# simulate a trend by plotting the cumulative mean of sepal length 
# to show how the average changes as we add more samples.
iris_df['cumulative_sepal_length_mean'] = iris_df['sepal length (cm)'].expanding().mean()
plt.figure(figsize=(10, 6))
plt.plot(iris_df['cumulative_sepal_length_mean'], color='teal', linewidth=2)
plt.title('Cumulative Mean of Sepal Length Over Samples')
plt.xlabel('Sample Index')
plt.ylabel('Cumulative Mean Sepal Length (cm)')
plt.show()
```
